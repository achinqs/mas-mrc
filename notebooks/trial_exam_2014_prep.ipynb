Assume you have to coordinate frames $A$ and $B$. The offset of the coordinate origin of $B$ relative to $A$ is $\mathbf{off} = <-1, 2.36, 9>$. Addtionally, $B$ is rotated by $37^\circ$ around $A$'s X axis. Construct the complete transformation matrix using numpy arrays. Transform the point $^A{P_1} = <3, -4.22, -0.8931>$ from $A$ to $B$.

```python
# Answer 2.1
from sympy.physics.mechanics import ReferenceFrame


A = ReferenceFrame('A')
B = A.orientnew('B','Axis',[radians(37) ,A.x])
offset = np.array([[-1], [2.36], [9]], dtype=float64)
R = np.array(np.asarray(A.dcm(B)), dtype=float64)
R = np.hstack([R, offset])
R = np.vstack([R, [0, 0, 0, 1]])

P_A = np.array([[3], [-4.22], [-0.8931], [1]], dtype=float64)
P_B = R.dot(P_A)
print P_B
```

Assume $A$ is a real, symmetric matrix with the unit eigenvectors $v_1$ and $v_2$ and corresponding eigenvalues $\lambda_1$ and $\lambda_2$ with $\lambda_1 \neq \lambda_2$. What is the dot product of $v_1$ and $v_2$? Give a detailed explanation (i.e. proof) of your answer.

$(\lambda_1 - \lambda_2)<v_1, v_2> = 0, \lambda_1 \neq \lambda_2 \Rightarrow <v_1, v_2> = 0 \Rightarrow v_1 \perp v_2$
