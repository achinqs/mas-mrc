### Assignment 3.1 <span style="color: green;">[L1]</span>

Take the following matrix and obtain it's eigenvalues:

$A_0 = \begin{bmatrix}1.35264537e+00 & -2.83012728e-03 & 3.75980444e-01 \\ -2.83012728e-03 & 1.00491214e-02 & -2.40593554e-03\\ 3.75980444e-01 & -2.40593554e-03 & 4.51412148e+00\end{bmatrix}$

Now obtain the QR decomposition of $A_0$ and implement the following algorithm:

1. $A_{k+1} = {Q_k^{-1}} \cdot A_k \cdot Q_k$

...and run it for at least 1500 iterations of $k$. What do you notice about the resulting matrix $A_{1500}$?

```python
# Solution 3.1
# ...

import numpy as np
import numpy.linalg as LA

A = np.array([[  1.35264537e+00,  -2.83012728e-03,   3.75980444e-01],
              [ -2.83012728e-03,   1.00491214e-02,  -2.40593554e-03],
              [  3.75980444e-01,  -2.40593554e-03,   4.51412148e+00]])

print sorted(LA.eig(A)[0])

for i in xrange(6000):
    Q,_ = LA.qr(A)
    A = np.dot(np.dot(Q.T,A),Q)
    
print sorted(LA.eig(A)[0])
print "I notice the eigenvalues are the same."
print "EDIT - I used sorted and xrange from the video"
```

```python
# Solution 3.3
# ...

import numpy as np
import numpy.linalg as LA

import matplotlib.pyplot as plt

# According to the video I just watched I shouldn't do most of what I'm about to do:
# EDIT - It was worse than I thought..

data = np.load("shape001.npy")

x, y = data.T
A = np.vstack([x, np.ones(len(x))]).T
m, c = LA.lstsq(A, y)[0]


x1, y1 = data[np.where(y>m*x+c)[0],:].T
x2, y2 = data[np.where(y<m*x+c)[0],:].T

A1 = np.vstack([x1, np.ones(len(x1))]).T
m1, c1 = LA.lstsq(A1, y1)[0]

A2 = np.vstack([x2, np.ones(len(x2))]).T
m2, c2 = LA.lstsq(A2, y2)[0]

fig = plt.figure()
ax2 = fig.add_subplot(111)
ax2.scatter(x,y)
ax2.plot(x1, m1*x1 + c1, 'r')
ax2.plot(x2, m2*x2 + c2, 'r')

data = np.load("shape002.npy")

x, y = data.T
A = np.vstack([x, np.ones(len(x))]).T
m, c = LA.lstsq(A, y)[0]


x1, y1 = data[np.where(y>m*x+c)[0],:].T
x2, y2 = data[np.where(y<m*x+c)[0],:].T

A1 = np.vstack([x1, np.ones(len(x1))]).T
m1, c1 = LA.lstsq(A1, y1)[0]

A2 = np.vstack([x2, np.ones(len(x2))]).T
m2, c2 = LA.lstsq(A2, y2)[0]


fig = plt.figure()
ax2 = fig.add_subplot(111)
ax2.scatter(x,y)
ax2.plot(x1, m1*x1 + c1, 'r')
ax2.plot(x2, m2*x2 + c2, 'r')

data = np.load("shape003.npy")

x, y = data.T
A = np.vstack([x, np.ones(len(x))]).T
m, c = LA.lstsq(A, y)[0]


x1, y1 = data[np.where(y>m*x+c)[0],:].T
x2, y2 = data[np.where(y<m*x+c)[0],:].T

A1 = np.vstack([x1, np.ones(len(x1))]).T
m1, c1 = LA.lstsq(A1, y1)[0]

A2 = np.vstack([x2, np.ones(len(x2))]).T
m2, c2 = LA.lstsq(A2, y2)[0]


fig = plt.figure()
ax3 = fig.add_subplot(111)
ax3.scatter(x,y)
ax3.plot(x1, m1*x1 + c1, 'r')
ax3.plot(x2, m2*x2 + c2, 'r')

plt.show()
```
