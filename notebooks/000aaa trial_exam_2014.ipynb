Assume you have to coordinate frames $A$ and $B$. The offset of the coordinate origin of $B$ relative to $A$ is $\mathbf{off} = <-1, 2.36, 9>$. Addtionally, $B$ is rotated by $37^\circ$ around $A$'s X axis. Construct the complete transformation matrix using numpy arrays. Transform the point $^A{P_1} = <3, -4.22, -0.8931>$ from $A$ to $B$.

```python
# Answer 2.1
# ...
import sympy.physics.mechanics as mechanics

A = mechanics.ReferenceFrame("A")

offset = np.array([-1, 2.36, 9])
offset.shape = (3,1)

extra_row = np.array([0, 0, 0, 1])
extra_row.shape = (1,4)

aP_1 = np.array([3, -4.22, -0.8931])
aP_1.shape = (3,1)

B = A.orientnew('B', 'Axis', (radians(37.0), A.x))

R = A.dcm(B)
R = np.array(np.asarray(R), dtype=np.float64)

R = np.hstack((R, offset))
R = np.vstack((R, extra_row))

aP_1 = np.vstack((aP_1, [1]))

print R

bP_1 = dot(R,aP_1)


print bP_1
```

Assume the general form of the step response is:

$c(t) = A + B \cdot \mathit{e}^{-6 \cdot t} \cdot cos(19.08 \cdot t + \phi)$

Take the following transfer function of a system, determine it's step response and obtain reasonable values for the general form of the step response using scipy.optimize.curve_fit:
    
$G_1(s) = \frac{400}{s^2 + 12 \cdot s + 400}$

```python
# Answer 2.2
# ...

import control.matlab as ml

from scipy.optimize import curve_fit

def func(t, A, B, phi):
    return A + B*np.exp(-6*t) * np.cos(19.08*t + phi)

G1 = ml.tf([400],[1, 12, 400])

y, t = ml.step(G1)

plt.figure(1)
plt.plot(t, y, 'r-')

popt, pcov = curve_fit(func, t, y)

A, B, phi = popt

yf = func(t, A, B, phi)

plt.plot(t, yf, 'b.')
plt.show()
```
