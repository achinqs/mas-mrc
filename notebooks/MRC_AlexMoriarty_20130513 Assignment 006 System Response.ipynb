Take the following three transfer functions and simulate their step responses:

<ul style="list-style-type: none;">
<li>$T_1 = \frac{73.626}{(s+3)(s^2 + 4s + 24.542)}$</li>
<li><p>&nbsp;</p></li>
<li>$T_2 = \frac{245.42}{(s+10)(s^2 + 4s + 24.542)}$<br /></li>
<li><p>&nbsp;</p></li>
<li>$T_3 = \frac{490.84}{(s+20)(s^2 + 4s + 24.542)}$<br /></li>
</ul>

Which of these third order systems can be best approximated using $T_{approx} = \frac{24.542}{s^2 + 4s + 24.542}$? **Why?**

All of these systems can be approximated using $T_{approx} = \frac{24.542}{s^2 + 4s + 24.542}$, they have just had the $\frac{1}{s+n}  $, where $n$ = $3$ $10$ or $20$ factored out. leaving $24.542$ in the denominator. Since $\frac{1}{s+n} $,  is the laplace transform of an exponential $e^{-nt}$ the larger the n the lesser the effect will be. Therefor the system  which is approximated the best is $T_3 = \frac{490.84}{(s+20)(s^2 + 4s + 24.542)}$.

As $n$ in $\frac{24.542n}{(s+n)(s^2 + 4s + 24.542)}$ grows, the function will approach $T_{approx} = \frac{24.542}{s^2 + 4s + 24.542}$.

```python
T1 = 73.626/((s+3)*(s**2+4*s+24.542))
T2 = 245.42/((s+10)*(s**2+4*s+24.542))
T3 = 490.84/((s+20)*(s**2+4*s+24.542))
#n = 100
#T4 = (24.542*n)/((s+n)*(s**2+4*s+24.542))
Ta = 24.542/(s**2+4*s+24.542)

stepResponse(T1,np.arange(0.0,3.0,0.01))
stepResponse(T2,np.arange(0.0,3.0,0.01))
stepResponse(T3,np.arange(0.0,3.0,0.01))
#stepResponse(T4,np.arange(0.0,3.0,0.01))
stepResponse(Ta,np.arange(0.0,3.0,0.01))
plt.show()
```
